Aggregate Function.....
#Calculating sum of CO column with respect to city.

```


CO_Sum = df1.groupby("AQI_Bucket",as_index=False).agg(
    {"CO": np.sum, "City": pd.Series.nunique})
```

```
CO_Sum.head()
```

Calculating Avg of CO by grouping w.r.t. AQI_BUCKET

```
.
CO_Avg = df1.groupby("AQI_Bucket",as_index=False).agg(
    {"CO": np.mean, "City": pd.Series.nunique})
```

```
CO_Avg=spark.createDataFrame(CO_Avg)
```

```
CO_Avg.head()
```

Calculating Avg of NH3 by grouping it w.r.t to AQI_BUCKET.

```

NH3_Avg = df1.groupby("AQI_Bucket",as_index=False).agg(
    {"NH3": np.mean, "City": pd.Series.nunique})
```

```
NH3_Avg=spark.createDataFrame(NH3_Avg)
```

```
NH3_Avg.tail()
```

Calculating Avg of AQI by grouping it w.r.t to AQI_BUCKET

```

AQI_Avg = df1.groupby("AQI_Bucket",as_index=False).agg(
    {"AQI": np.mean})
```

```
AQI_Avg.head()
```

Calculating Avg of AQI w.r.t year.

```

year = df1.groupby("year",as_index=False).agg(
    {"AQI": np.mean, "City": pd.Series.nunique})
```

Calculating Avg of AQI w.r.t month.

```

month = df1.groupby("month",as_index=False).agg(
    {"AQI": np.mean, "City": pd.Series.nunique})
```
